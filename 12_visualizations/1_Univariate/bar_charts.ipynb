```python
pkmn_types = pokemon.melt(id_vars = ['id','species'], 
                          value_vars = ['type_1', 'type_2'], 
                          var_name = 'type_level', value_name = 'type').dropna()
pkmn_types.head()
```

```python
# relative Frequency 
count_max = pkmn_types['type_level'].value_counts().max()
n_count = pkmn_types.shape[0]
max_pro = count_max / n_count
```

```python
#generating the tick mark locations 
tick_props = np.arange(0, max_pro, 0.05)
tick_names =  ['{:0.2f}'.format(v) for v in tick_props]
# creating the plot 
type_counts = pkmn_types['type'].value_counts()
type_order = type_counts.index


sns.countplot(data= pkmn_types,x = 'type' )
plt.yticks(tick_props * n_count, tick_names)
plt.ylabel('proportion')
```
